```python
import matplotlib.pyplot as plt
```

```python
with open("tensorboard/han_voc_50/Train_Loss.txt", "r") as f:
    tb = list(map(lambda x:float(x.split(",")[0]), f.readlines()))
with open("tensorboard/han_voc_100/Train_Loss.txt", "r") as f:
    l1 = list(map(lambda x:float(x.split(",")[0]), f.readlines()))
with open("tensorboard/han_voc_200/Train_Loss.txt", "r") as f:
    l2 = list(map(lambda x:float(x.split(",")[0]), f.readlines()))
with open("tensorboard/han_voc_300/Train_Loss.txt", "r") as f:
    l3 = list(map(lambda x:float(x.split(",")[0]), f.readlines()))
with open("tensorboard/han_voc_300p/Train_Loss.txt", "r") as f:
    l4 = list(map(lambda x:float(x.split(",")[0]), f.readlines()))
plt.plot(range(len(tb)), tb,label = "d = 50, glove.6B")
plt.plot(range(len(l1)), l1,label = "d = 100, glove.6B")
plt.plot(range(len(l2)), l2,label = "d = 200, glove.6B")
plt.plot(range(len(l3)), l3,label = "d = 300, glove.6B")
plt.plot(range(len(l4)), l4,label = "d = 300, glove.840B")
plt.legend()
plt.xlabel("# of batches")
plt.ylabel("Accuracy")
plt.title("Accuracy in each Batch, Training set")
plt.show()
```
